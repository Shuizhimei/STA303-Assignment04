```python
import matplotlib.pyplot as plt
import numpy as np
```

### Alpha with different datasets

```python
# SAPS, Split predictor
# MNIST, GTSRB, DTD, Flowers102
alpha_values = [0.01, 0.05, 0.1, 0.2, 0.5]
dataset1 = [2.27,1.29,1.16,0.97,0.55]
dataset2 = [41.59,39.28,36.48,33.29,25.39]
dataset3 = [46.44,44.77,42.59,38.11,23.36] 
dataset4 = [76.72,53.09,40.10,24.71, 7.33]

plt.figure(figsize=(8, 6))

plt.plot(alpha_values, dataset1, label='MNIST', marker='o')
plt.plot(alpha_values, dataset2, label='GTSRB', marker='s')
plt.plot(alpha_values, dataset3, label='DTD', marker='^')
plt.plot(alpha_values, dataset4, label='Flowers102', marker='d')

plt.xlabel('Alpha')
plt.ylabel('Set Size')
plt.title('Alpha vs Set Size for Different Datasets')
plt.legend()
plt.grid(True)

plt.savefig('./result/alpha_plot.png',dpi=500)
plt.show()
```

### Score with Set Size & Coverage Rate

```python
# 0.05, ClassWisePredictor
def score_plot(rate_list, size_list, name):
    fig, ax1 = plt.subplots()
    score_labels = ["THR", "APS", "RAPS", "SAPS"]

    color = 'tab:blue'
    ax1.set_xlabel('Score')
    ax1.set_ylabel('Set size', color=color)
    ax1.bar(score_labels, size_list, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.set_ylabel('Coverage rate', color=color)
    ax2.plot(score_labels, rate_list, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    plt.savefig(name,dpi=400)

    plt.show()
```

```python
# MNIST
rate_list = [0.9424, 0.948, 0.9444, 0.9438]
size_list = [1.059, 1.4156, 1.1004, 1.2858]
score_plot(rate_list,size_list,"./result/score_MNIST.png")
```

```python
# DTD
rate_list = [0.40, 0.40, 0.74, 0.98]
size_list = [16.52, 16.59, 33.97, 45.53]
score_plot(rate_list,size_list,"./result/score_DTD.png")
```

```python
# GTSRB
rate_list = [0.074, 0.058, 0.244, 0.949]
size_list = [1.013, 1.027, 12.08, 40.47]
score_plot(rate_list,size_list,"./result/score_GTSRB.png")
```

```python
# Flowers102
rate_list = [0.96, 0.957, 0.96, 0.96]
size_list = [69.01, 70.70, 69.97, 69.86]
score_plot(rate_list,size_list, "./result/score_Flowers.png")
```

### Predictor with Set Size & Coverage Rate

```python
# 0.05, SAPS
def predictor_plot(rate_list, size_list,name):
    fig, ax1 = plt.subplots()
    predictor_labels = ["SplitPredictor", "ClassWisePredictor", "ClusterPredictor"]

    color = 'tab:blue'
    ax1.set_xlabel('Predictor')
    ax1.set_ylabel('Set size', color=color)
    ax1.bar(predictor_labels, size_list, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.set_ylabel('Coverage rate', color=color)
    ax2.plot(predictor_labels, rate_list, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    plt.savefig(name,dpi=500)

    plt.show()
```

```python
# MNIST
rate_list = [0.9478,0.9438, 0.941]
size_list = [1.2914,1.2858, 1.277]
predictor_plot(rate_list,size_list,"./result/predictor_MNIST.png")
```

```python
# DTD
rate_list = [0.967, 0.979, 0.967]
size_list = [44.98, 45.53, 44.85]
predictor_plot(rate_list,size_list,"./result/predictor_DTD.png")
```

```python
# GTSRB
rate_list = [0.950, 0.949, 0.947]
size_list = [39.28, 40.47, 38.91]
predictor_plot(rate_list,size_list,"./result/predictor_GTSRB.png")
```

```python
# Flowers102
rate_list = [0.944, 0.959, 0.945]
size_list = [53.09, 69.86, 53.46]
predictor_plot(rate_list,size_list,"./result/predictor_Flowers.png")
```
